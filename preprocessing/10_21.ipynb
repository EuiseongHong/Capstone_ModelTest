```python
from ultralytics import YOLO
```

```python
#기본 모델

model = YOLO("yolo11n.pt")
results = model.track(
    source="./videos/test_video.mp4",
    tracker="bytetrack.yaml",
    conf=0.4,
    persist=True,
    save=True,          #bounding box 그려진 비디오 저장
    save_txt=True,      #각 프레임마다의 bounding box 정보
    show=False,
    stream=False,
    name="test_eval"  #저장 폴더명
)
```

```python
#######train 한 모델 이용하기
model = YOLO("better.pt")
results = model.track(
    source="./videos/crash_ver2.mov",
    tracker="bytetrack.yaml",
    conf=0.4,
    persist=True,
    save=True,          #bounding box 그려진 비디오 저장
    save_txt=True,      #각 프레임마다의 bounding box 정보
    show=False,
    stream=False,
    name="test_crash_ver2"  #저장 폴더명
)
```
